#Importando Dados

````python
import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados.head()
````

````python
dados.columns = dados.columns.str.replace('-','')
dados.head()
````

````python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
````

````python
dados_resultados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true')
dados_resultados.head()
````

````python
contagem_moa = dados_resultados.drop('id', axis=1).sum().sort_values(ascending=False)
contagem_moa
````

````python
dados_resultados['n_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1)
````

````python
dados_resultados['ativo_moa'] = (dados_resultados['n_moa'] != 0)
dados_resultados.head()
````

````python
 dados_combinados = pd.merge(dados, dados_resultados[['id','n_moa', 'ativo_moa']], on='id')
 dados_combinados.head()
````

O deafio propóem que ao invés de tentarmos prever se exitem ou não MoAs ativos, que nossa análise seja feita em prever quantos MoAs serão ativados.
Primeiro iremos utilizar o primeiro método da aula, a regressão logística.

````python
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['n_moa'] 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_rlogistica = LogisticRegression(max_iter=10000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
````

Obtivemos um resultado de 58% de acerto, o que parece ser baixo, e como na aula iremos testar com o método ```DummyClassifier('most frequent')```.

````python
modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
````

Assim como na aula, os resultados da primeira análise não ficam muito distantes do nosso teste "Dummy", de 52%, portanto utilizaremos um outro método, nos baseando no teste dummy para averiguar se nossos resultados são bons ou ruins.
Utilizaremos o método de ```RandomForestClassifier``` proposto na aula, apenas alterando o objeto de análise de ativo ou não, para a quantidade ativa.

````python
x = dados_combinados.drop(['id', 'n_moa', 'ativo_moa', 'composto'], axis=1)
x = pd.get_dummies(x, columns=['tratamento', 'dose', 'tempo'])
y = dados_combinados['n_moa'] 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_randomforest = RandomForestClassifier()
modelo_randomforest.fit(x_treino, y_treino)
modelo_randomforest.score(x_teste, y_teste)
````

Através dessa análise pudemos verificar que não houve grande avanço em relação a predição. O modelo acertou aproximadamente 63% das quantidades de MoAs.

##Desafio 05: Escolha a droga mais utilizada e tente criar um modelo para prever está label. 


Neste desafio, usaremos o modelo de machine learning para tentar prever se o composto é o mais utilizado ou não, inicialmente em relação ao controle.

````python
dados_combinados['composto'].value_counts().head(2)
````

````python
#Por estudos anteriores, sabemos que 'cacb2b860' é o composto de controle, portanto o composto mais utilizado é '87d714366'
compostos_analise = ['87d714366', 'cacb2b860']
dados_analise = dados_combinados.query('composto in @compostos_analise')
````

````python
dados_analise
````

Se incluirmos na analise as colunas ```'tratamento'```, ```'ativo_moa'``` e ```'n_moa'``` estaremos "vazando dados" como dito em aula, pois essas colunas podem facilmente diferenciar o composto do controle, portanto serão excluidas da analise.  
Utilizamos a função ```pd.get.dummies()``` para criar as colunas que transformam as caracteristicas qualitativas (D1 ou D2, por exemplo), dos experimentos em 0's e 1's, para que possa ser entendido melhor pelo método.

````python
x = dados_analise.drop(['id','composto','tratamento', 'n_moa', 'ativo_moa',], axis=1)
x = pd.get_dummies(x, columns=['dose', 'tempo'])
y = dados_analise['composto']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_randomforest = RandomForestClassifier()
modelo_randomforest.fit(x_treino, y_treino)
modelo_randomforest.score(x_teste, y_teste)
````

Pelo resultado, nosso modelo acertou 100% das vezes, se o composto era diferente do controle, o que nos leva a inferir que essa distinção é muito perceptivel.  
Portanto, iremos então testar o modelo para todo o DF, porém como possuem compostos os quais foram testados apenas 1 vez, e o minimo para o modelo funcionar é 2, porém iremos excluir os compostos com menos de 10 experimentos, para reduzir o tempo de execução.

````python
# Primeiro adquirimos as informações de quantidades:
value_counts = dados_combinados['composto'].value_counts()

# Selecionamos os valores a serem removidos:
to_remove = value_counts[value_counts < 10].index

#Então removemos os dados:
dados_teste = dados_combinados[~dados_combinados.composto.isin(to_remove)]

dados_teste['composto'].value_counts()
````

````python
x = dados_teste.drop(['id','composto','tratamento', 'n_moa', 'ativo_moa',], axis=1)
x = pd.get_dummies(x, columns=['dose', 'tempo'])
y = dados_teste['composto']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_randomforest = RandomForestClassifier()
modelo_randomforest.fit(x_treino, y_treino)
modelo_randomforest.score(x_teste, y_teste)
````

O modelo chegou a um resultado de aproximadamente 71% de acerto em relação ao composto utilizado, para efeito de comparação, iremos mais uma vez utilizar o método ```DummyClassifier('most_frequent')```

````python
modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
````

O método ```RandomForest``` nos trouxe um resultado maior que o dobro em relação ao ```DummyClassifier```.
